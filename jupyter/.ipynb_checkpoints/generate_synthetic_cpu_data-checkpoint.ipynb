```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

def generate_synthetic_cpu_data(num_samples, noise_level=0.05, spike_freq=50, random_seed=42):
    """
    Generate synthetic CPU usage data.
    
    Parameters:
    - num_samples: int, number of data points to generate.
    - noise_level: float, standard deviation of the noise.
    - spike_freq: int, frequency of artificial spikes in the data.
    - random_seed: int, seed for reproducibility.
    
    Returns:
    - DataFrame containing synthetic CPU usage data.
    """
    np.random.seed(random_seed)
    
    # Create a time series with a slight upward trend and some sinusoidal variation
    time = np.arange(num_samples)
    baseline = np.sin(2 * np.pi * time / 180) + time / 200
    
    # Add random spikes
    spikes = np.zeros(num_samples)
    for i in range(num_samples // spike_freq):
        spike_index = np.random.randint(0, num_samples)
        spikes[spike_index] += np.random.rand() * 10  # Spike by up to 10% CPU usage
    
    # Add noise
    noise = np.random.normal(scale=noise_level, size=num_samples)
    
    # Combine all components to form the synthetic CPU usage data
    cpu_usage = baseline + noise + spikes
    cpu_usage = np.clip(cpu_usage, 0, 100)  # Limit values to between 0% and 100%
    
    # Create a DataFrame
    df = pd.DataFrame({'CPU_Usage': cpu_usage, 'Time': pd.date_range(start='20210101', periods=num_samples, freq='T')})
    df.set_index('Time', inplace=True)
    
    return df

def preprocess_data(df, sequence_length):
    """
    Create sequences and labels for LSTM training from time series data.
    
    Parameters:
    - df: DataFrame, the data frame containing the time series data.
    - sequence_length: int, number of time steps in each input sequence.
    
    Returns:
    - X: np.array, input data for LSTM (sequences).
    - y: np.array, labels for LSTM (next value following each sequence).
    """
    # Normalize the CPU usage data
    scaler = MinMaxScaler(feature_range=(0, 1))
    df['CPU_Usage'] = scaler.fit_transform(df[['CPU_Usage']])
    
    # Create sequences
    X = []
    y = []
    for i in range(len(df) - sequence_length):
        X.append(df['CPU_Usage'].iloc[i:i + sequence_length].values)
        y.append(df['CPU_Usage'].iloc[i + sequence_length])
    
    return np.array(X), np.array(y)
```

```python
import numpy as np
import pandas as pd

# Generate synthetic CPU usage data
num_samples = 1000  # Number of data points to generate
noise_level = 0.05  # Standard deviation of the noise
spike_freq = 50     # Frequency of artificial spikes
random_seed = 42    # Seed for reproducibility

cpu_data = generate_synthetic_cpu_data(num_samples, noise_level, spike_freq, random_seed)
```

```python
import matplotlib.pyplot as plt

# Plot CPU usage data
plt.figure(figsize=(10, 6))
plt.plot(cpu_data.index, cpu_data['CPU_Usage'], color='blue', label='CPU Usage')
plt.xlabel('Time')
plt.ylabel('CPU Usage')
plt.title('Synthetic CPU Usage Data')
plt.legend()
plt.show()
```
