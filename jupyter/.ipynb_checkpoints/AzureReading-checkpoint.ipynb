```python
# Importing necessary libraries
import math
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import keras
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AzureReadings.csv')
df['timestamp'] =  pd.to_datetime(df['timestamp'])
df = df.set_index('timestamp')
df.head()
```

```python
dataset = df.values
dataset = dataset.astype('float32')
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
```

```python
train_size = int(len(dataset) * 0.8)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size, :], dataset[train_size:len(dataset), :]
print(len(train), len(test))
```

```python
def create_training_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), :3]
        dataX.append(a)
        dataY.append(dataset[i + look_back, :])
    return np.array(dataX), np.array(dataY)
```

```python
look_back = 5
trainX, trainY = create_training_dataset(train, look_back=look_back)
testX, testY = create_training_dataset(test, look_back=look_back)
```

```python
trainX.shape
```

```python
model = keras.models.Sequential()
model.add(keras.layers.LSTM(128, input_shape=(trainX.shape[1], trainX.shape[2])))
model.add(keras.layers.Dense(3))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
# callback = keras.callbacks.EarlyStopping(monitor='loss', patience=3)
history = model.fit(trainX, trainY, validation_split=0.2,epochs=20, batch_size=64, verbose=2)
```

```python
model.evaluate(testX, testY)
```

```python
model.metrics_names
```

```python
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
```

```python
trainY = scaler.inverse_transform(trainY)
trainPredict = scaler.inverse_transform(trainPredict)
testY = scaler.inverse_transform(testY)
testPredict = scaler.inverse_transform(testPredict)
```

```python
trainScore = math.sqrt(mean_squared_error(trainY[:], trainPredict[:]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testY[:], testPredict[:]))
print('Test Score: %.2f RMSE' % (testScore))
```

```python
from sklearn.metrics import mean_absolute_error
trainScore = (mean_absolute_error(trainY[:], trainPredict[:]))
print('Train Score: %f MAE' % (trainScore))
testScore = math.sqrt(mean_absolute_error(testY[:], testPredict[:]))
print('Test Score: %f MAE' % (testScore))
```

```python
trainScore2 = np.mean(np.abs(trainPredict - trainY)/np.abs(trainY))
print('Train Score: %f MAPE' % (trainScore2))
testScore2 = np.mean(np.abs(testPredict - testY)/np.abs(testY))
print('Test Score: %f MAPE' % (testScore2))
```

```python
def calculate_precision(actual, predicted, tolerance_percent=10):
    """Calculate the percentage of predictions within the specified tolerance level of the actual values."""
    # Calculate the absolute percentage error
    errors = np.abs(actual - predicted) / actual
    # Calculate the percentage of predictions within the tolerance level
    within_tolerance = errors < (tolerance_percent / 100)
    # Compute the precision as a percentage
    precision = np.mean(within_tolerance) * 100
    return precision

# Define the tolerance level
tolerance_percent = 10

# Calculate precision for training and testing data
train_precision = calculate_precision(trainY, trainPredict, tolerance_percent)
test_precision = calculate_precision(testY, testPredict, tolerance_percent)

print(f"Training Data Precision: {train_precision:.2f}% within {tolerance_percent}% tolerance")
print(f"Testing Data Precision: {test_precision:.2f}% within {tolerance_percent}% tolerance")
```
